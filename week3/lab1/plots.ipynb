```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("second.csv", delimiter='\t')
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Target"], df["Standart"], color='#4CAF50', edgecolor='black')
plt.title('Standart Dice', fontsize=18, color='#333333')
plt.ylabel(r'Вероятность попадания $\omega$', fontsize=14, color='#555555')
plt.xlabel('Значение X', fontsize=14, color='#555555')
plt.xlim([1,100])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Target"], df["ThreeDice"], color='#FF9800', edgecolor='black')
plt.title('Three Dice', fontsize=18, color='#333333')
plt.ylabel(r'Вероятность попадания $\omega$', fontsize=14, color='#555555')
plt.xlabel('Значение X', fontsize=14, color='#555555')
plt.xlim([2,19])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Target"], df["Penalty"], color='#2196F3', edgecolor='black', label="Penalty", alpha=0.7)
plt.bar(df["Target"], df["Bonus"], color='#F44336', edgecolor='black', label="Bonus", alpha=0.5)
plt.title('Penalty & Bonus Dice', fontsize=18, color='#333333')
plt.ylabel(r'Вероятность попадания $\omega$', fontsize=14, color='#555555')
plt.xlabel('Значение X', fontsize=14, color='#555555')
plt.xlim([1,100])
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
